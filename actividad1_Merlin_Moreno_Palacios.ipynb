```python
# importar las libreridas que se requieren para el proyecto
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('shoes_sales_dataset.csv')
df.head(1000
        )
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
# se convierte la columna "Date" a formato de fecha, y se maneja cualquier error que pueda surgir durante la conversión.
df.info()
```

```python
df.describe() #
```

```python
# Se obtiene el valor mínimo y máximo de cada columna, 
# para conocer el rango de fechas, precios, unidades vendidas e ingresos.

date1 = df["Date"].min()
date2 = df["Date"].max()
print(f"La fecha más antigua es: {date1}")
print(f"La fecha más reciente es: {date2}")
p1 = df["Price_USD"].min()
p2 = df["Price_USD"].max()
print(f"El precio más bajo es: {p1}")
print(f"El precio más alto es: {p2}")
u1 = df["Units_Sold"].min()
u2 = df["Units_Sold"].max()
print(f"La cantidad de unidades vendidas más baja es: {u1}")
print(f"La cantidad de unidades vendidas más alta es: {u2}")
R1 = df["Revenue_USD"].min()
R2 = df["Revenue_USD"].max()        
print(f"El ingreso más bajo es: {R1}")
print(f"El ingreso más alto es: {R2}")
```

```python
df.head(2)
```

```python
# grafica de barra para mostrar las ventas por cada tipo de marca de zapatos
brand_sales = df.groupby("Brand")["Units_Sold"].sum().reset_index()
plt.figure(figsize=(7, 5))
plt.bar(brand_sales["Brand"], brand_sales["Units_Sold"])
plt.xlabel("Marca")
plt.ylabel("Unidades Vendidas")
plt.title("Ventas por Marca")
plt.xticks(rotation=45)
plt.show()
```

```python
# grafica de barra para mostrar el precio promedio de las ventas por cada tipo de marca de zapatos
brand_prices = df.groupby("Brand")["Price_USD"].mean().reset_index()
plt.figure(figsize=(7, 5))
plt.bar(brand_prices["Brand"], brand_prices["Price_USD"])
plt.xlabel("Marca")
plt.ylabel("Precio Promedio (USD)")
plt.title("Ventas por Marca")
plt.xticks(rotation=45)
plt.show()
```

##### La marca con más unidades vendidas fue SKECHERS. Los precios promedio por marca se encuentran alrededor de 137 euros , lo cual es consistente con la tabla de descripcion de datos numéricos

```python
# graficas de ventas mensuales, ingresos mensuales 
df["Month"] = df["Date"].dt.to_period("M")
monthly_sales = df.groupby("Month")["Revenue_USD"].sum().reset_index()
plt.figure(figsize=(8, 6))
plt.bar(monthly_sales["Month"].astype(str), monthly_sales["Revenue_USD"])
plt.xlabel("Mes")
plt.ylabel("Ingresos Totales (USD)")
plt.title("Ventas Mensuales")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

##### Los meses con mayor venta son octubre, en primer lugar, con ventas superiores a 140.000 USD, seguido de julio y noviembre.

```python
# grafica de torta de ventas por canal de venta
channel_sales = df.groupby("Sales_Channel")["Units_Sold"].sum().reset_index()
plt.figure(figsize=(3, 3))
plt.pie(channel_sales["Units_Sold"], labels=channel_sales["Sales_Channel"], autopct="%1.1f%%", startangle=140)
plt.title("Porcentaje de Ventas por Canal")
plt.axis("equal")
plt.show()

```
