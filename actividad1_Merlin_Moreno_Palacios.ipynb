```python
# importar las libreridas que se requieren para el proyecto
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('shoes_sales_dataset.csv')
df.head(1000
        )
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"], errors='coerce')
# se convierte la columna "Date" a formato de fecha, y se maneja cualquier error que pueda surgir durante la conversión.
df.info()
```

```python
df.describe() #
```

```python
# Se obtiene el valor mínimo y máximo de cada columna, 
# para conocer el rango de fechas, precios, unidades vendidas e ingresos.

date1 = df["Date"].min()
date2 = df["Date"].max()
print(f"La fecha más antigua es: {date1}")
print(f"La fecha más reciente es: {date2}")
p1 = df["Price_USD"].min()
p2 = df["Price_USD"].max()
print(f"El precio más bajo es: {p1}")
print(f"El precio más alto es: {p2}")
u1 = df["Units_Sold"].min()
u2 = df["Units_Sold"].max()
print(f"La cantidad de unidades vendidas más baja es: {u1}")
print(f"La cantidad de unidades vendidas más alta es: {u2}")
R1 = df["Revenue_USD"].min()
R2 = df["Revenue_USD"].max()        
print(f"El ingreso más bajo es: {R1}")
print(f"El ingreso más alto es: {R2}")
```
